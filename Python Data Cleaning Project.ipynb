```python
# import our libraries
import pandas as pd
import numpy as np
import openpyxl


# Loading the dirty excel data into pandas
df_customer_list = pd.read_excel(r'C:\Users\GBOGBO ELIJAH\Documents\Pandas Tutorial\Customer Call List.xlsx')
df_customer_list


# Checking for duplicates in the data
df_customer_list.duplicated()


# Removing duplicates from the data
df_customer_list = df_customer_list.drop_duplicates()
df_customer_list


# Removing a column that is not useful
df_customer_list = df_customer_list.drop(columns= 'Not_Useful_Column')
df_customer_list


# Cleaning the Last_Name column by stripping dirty characters (./_) from the string values
df_customer_list['Last_Name'] = df_customer_list['Last_Name'].str.strip('[._/]')
df_customer_list





# Standardizing the Phone_Number column to the correct phone number format


# Returning the Phone_Number values to only digits removing incorrect characters
df_customer_list['Phone_Number'] = df_customer_list['Phone_Number'].str.replace(pat= '[^a-zA-Z0-9]', repl= '', regex= True)

# Converting each number in the Phone_Number column to a string datatype
df_customer_list['Phone_Number'] = df_customer_list['Phone_Number'].apply(lambda x: str(x))

# Re-formatting the string phone numbers to the correct format
df_customer_list['Phone_Number'] = df_customer_list['Phone_Number'].apply(lambda x: x[:3] + '-' + x[3:6] + '-' + x[6:])

# Replacing incorrect values in the Phone_Number column with blank spaces
df_customer_list['Phone_Number'] = df_customer_list['Phone_Number'].replace(to_replace= 'nan--', value= '')
df_customer_list['Phone_Number'] = df_customer_list['Phone_Number'].replace(to_replace= 'Na--', value= '')
df_customer_list






# Splitting the Address column into 3 columns to provide a more descriptive data
df_customer_list[['Street_Address', 'State', 'Zip_Code']] = df_customer_list['Address'].str.split(',', n= 2, expand= True)
df_customer_list



# Removing the Address Column
df_customer_list = df_customer_list.drop(columns= 'Address')
df_customer_list



# Replacing the 'Y' and 'N' with 'Yes' and 'No' respectively in the Paying Customer column
df_customer_list['Paying Customer'] = df_customer_list['Paying Customer'].replace(to_replace= 'Y', value= 'Yes')
df_customer_list['Paying Customer'] = df_customer_list['Paying Customer'].replace(to_replace= 'N', value= 'No')
df_customer_list


# Replacing the 'Y' and 'N' with 'Yes' and 'No' respectively in the Do_Not_Contact column
df_customer_list['Do_Not_Contact'] = df_customer_list['Do_Not_Contact'].replace(to_replace= 'Y', value= 'Yes')
df_customer_list['Do_Not_Contact'] = df_customer_list['Do_Not_Contact'].replace(to_replace= 'N', value= 'No')
df_customer_list


# Replacing null and 'N/a' values with blank spaces
df_customer_list = df_customer_list.replace(to_replace= [np.nan, 'N/a'], value= '')
df_customer_list



# Removing the rows in which the Do_Not_Contact column value is 'Yes'
for val_x in df_customer_list.index:
    if df_customer_list.loc[val_x, 'Do_Not_Contact'] == 'Yes':
        df_customer_list.drop(index= val_x, axis= 0, inplace= True)

df_customer_list





# Replacing the blank spaces in the Do_Not_Contact column with 'No'
df_customer_list['Do_Not_Contact'] = df_customer_list['Do_Not_Contact'].replace(to_replace= '', value= 'No')
df_customer_list





# Removing the rows in which the Phone_Number column value is blank
for val_y in df_customer_list.index:
    if df_customer_list.loc[val_y, 'Phone_Number'] == '':
        df_customer_list.drop(index= val_y, axis= 0, inplace= True)


# Resetting the index values
df_customer_list.reset_index(drop= True, inplace= True)


# This the final cleaned call centre data
df_customer_list



```
